```
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```
df = pd.read_csv('monitoramento_tempo.csv')
```

```
df['data'] = pd.to_datetime(df['data'])
```

```
plt.figure(figsize=(15,8))
plt.title('Temperatura por tempo')
plt.plot(df['data'],df['temperatura'])
```

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['temperatura'])
eixo.set_title('Temperatura no tempo')
```

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['temperatura'])
eixo.set_title('Temperatura no tempo',fontsize=25)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Tempo',fontsize=20)
```

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['temperatura'], color='gray')
eixo.set_title('Temperatura no tempo',fontsize=25)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Tempo',fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
plt.show()
```

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'],df['temperatura'], color='g')
eixo.grid(True)
eixo.set_title('Temperatura no tempo',fontsize=25, pad=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Tempo',fontsize=20)

x1=df['data'][df['temperatura'].idxmax()]
y1=max(df['temperatura'])

eixo.annotate('Máximo', xy=(x1,y1), fontsize=15)

eixo.axhline(max(df['temperatura']),color='k',linestyle='--')
eixo.axhline(min(df['temperatura']),color='k',linestyle='--')

plt.show()
```

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'],df['temperatura'], color='g')
eixo.grid(True)
eixo.set_title('Temperatura no tempo',fontsize=25, pad=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Tempo',fontsize=20)

x1=df['data'][df['temperatura'].idxmax()]
y1=max(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmax()-7000]
y2=max(df['temperatura']-5)
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2),arrowprops=dict(facecolor='k'))

x1=df['data'][df['temperatura'].idxmin()]
y1=min(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmin()-7000]
y2=min(df['temperatura']+5)
eixo.annotate('Mínimo', xy=(x1,y1), fontsize=20, xytext=(x2,y2),arrowprops=dict(facecolor='k'))

eixo.axhline(max(df['temperatura']),color='k',linestyle='--')
eixo.axhline(min(df['temperatura']),color='k',linestyle='--')

plt.show()
```

```
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta','Quinta','Sexta','Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

index = range(len(temperatura_por_dia_da_semana))
colors = ['black', 'r', 'g','b','yellow', 'orange','magenta' ]

axis.bar(index,temperatura_por_dia_da_semana, color=colors)
axis.set_title('Temperatura media por dia da semana',fontsize=15, pad=10)
axis.set_xlabel('Dia da semana',fontsize=15)
axis.set_ylabel('Temperatura média', fontsize=15)

axis.set_xticks(index)
axis.set_xticklabels(nome_dias)

plt.show()
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

index = range(len(temperatura_por_dia_da_semana))
colors = ['black', 'r', 'g','b','yellow', 'orange','magenta' ]

axis.bar(index,temperatura_por_dia_da_semana, color=colors, edgecolor='black')
axis.set_title('Temperatura media por dia da semana',fontsize=15, pad=10)
axis.set_xlabel('Dia da semana',fontsize=15)
axis.set_ylabel('Temperatura média', fontsize=15)

axis.set_xticks(index)
axis.set_xticklabels(nome_dias)

plt.show()
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

axis.pie(temperatura_por_dia_da_semana,labels=temperatura_por_dia_da_semana.index)
axis.set_title('Temperatura media por dia da semana',fontsize=15, pad=10)

plt.show()
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

axis.pie(temperatura_por_dia_da_semana,labels=temperatura_por_dia_da_semana.index,
         autopct='%.2f%%')
axis.set_title('Temperatura media por dia da semana',fontsize=15, pad=10)

plt.show()
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

explode=[0.1,0,0,0,0,0.1,0.1]
axis.pie(temperatura_por_dia_da_semana,labels=temperatura_por_dia_da_semana.index,
         autopct='%.2f%%',explode=explode)
axis.set_title('Temperatura media por dia da semana',fontsize=15, pad=10)

plt.show()
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

explode=[0.1,0,0,0,0,0.1,0.1]
axis.pie(temperatura_por_dia_da_semana,labels=temperatura_por_dia_da_semana.index,
         autopct='%.2f%%',explode=explode, shadow=True)
axis.set_title('Temperatura media por dia da semana',fontsize=15, pad=10)

plt.show()
```
